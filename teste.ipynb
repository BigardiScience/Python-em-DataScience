```python
import datetime
# Definindo uma função
def minha_funcao():
    return "Olá mundo!"

# Definindo uma classe
class MinhaClasse:
    def __init__(self, nome):
        self.nome = nome

    def cumprimentar(self):
        return f"Olá, eu sou {self.nome}"

# Criando uma instância da classe
objeto = MinhaClasse("Python")

# Objetos para análise
item_1 = minha_funcao
item_2 = MinhaClasse
item_3 = objeto
item_4 = objeto.cumprimentar
item_5 = len  # função built-in
item_6 = datetime.datetime.now()

print(f'minha_funcao - {type(minha_funcao)}\n'
      ,f'MinhaClasse - {type(MinhaClasse)}\n'
      ,f'objeto - {type(objeto)}\n'
      ,f'Item_1 - {type(item_1)}\n'
      ,f'item_2 - {type(item_2)}\n'
      ,f'item_3 - {type(item_3)}\n'
      ,f'item_4 - {type(item_4)}\n'
      ,f'item_5 - {type(item_5)}\n'
      ,f'item_6 - {type(item_6)}')
```

```python
#  Estruturas complexas para o desafio final
from collections import defaultdict


desafio_a = [1, "texto", [2, 3], {"chave": "valor"}]
desafio_b = {"lista": [1, 2, 3], "tupla": (4, 5, 6), "funcao": len}
desafio_c = lambda x: x * 2
desafio_d = defaultdict(list)
desafio_e = enumerate([1, 2, 3])

class Veiculo:
    pass

desafio_f = Veiculo()
desafio_g = [Veiculo(), minha_funcao, {"nested": {"deep": True}}]
# desafio_a:
# Sua resposta:
print(f'desafio_a - {type(desafio_a)}\n',
      f'desafio_b - {type(desafio_b)}\n',
      f'desafio_c - {type(desafio_c)}\n',
      f'desafio_d - {type(desafio_d)}\n',
      f'desafio_e - {type(desafio_e)}\n',)

# Bonus: Qual é o tipo do primeiro elemento de desafio_a?
# desafio_a[0]:
# Sua resposta:
print(f'desafio_a primeiro elemento {type(desafio_a[0])}')
# Bonus: Qual é o tipo do valor da chave "funcao" em desafio_b?
# desafio_b["funcao"]:
# Sua resposta:
print(f'desafio_b elemento funcao {type(desafio_b["funcao"])}')
# desafio_c (função lambda):
# Sua resposta:
print(f'desafio_c elemento {type(desafio_c)}')
# desafio_d (defaultdict):
# Sua resposta:
print(f'desafio_d elemento {type(desafio_d)}')
# desafio_e (enumerate object):
# Sua resposta:
print(f'desafio_e elemento {type(desafio_e)}')
# desafio_f (instância de Veiculo):
# Sua resposta:
print(f'desafio_f elemento {type(desafio_f)}')
# desafio_g:
# Sua resposta:
print(f'desafio_g elemento {type(desafio_g)}')

# Super Bonus: Explore os tipos dos elementos dentro de desafio_g
# desafio_g[0]:
# Sua resposta:
print(f'desafio_g primeiro elemento {type(desafio_g[0])}')

# desafio_g[1]:
# Sua resposta:
print(f'desafio_g segundo elemento {type(desafio_g[1])}')
# desafio_g[2]:
# Sua resposta:
print(f'desafio_g terceiro elemento {type(desafio_g[2])}')
```
