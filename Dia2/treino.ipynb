```python
class pessoas:
    def __init__(self,nome,idade, end, rendaMensal):
        self.nome = nome
        self.idade = idade
        self.end = end
        self.mensal = rendaMensal

    def mostrarDadosPessoais(self):
        print(self.nome,self.idade,self.end)

    def mostrarRenda(self):
        print(self.mensal)

        
x = pessoas("Pedro", "21", "av 123", 1500)

x.mostrarDadosPessoais()
x.mostrarRenda()

 


```

```python
#extend = acrescentar multiplos valores em uma lista

frutas = ["Maça", "Banana"]

frutas.extend(["Acerola", "Pera"])  #Tuplas acrescenta com extend
frutas.remove("Maça")   #remove para remover um item
print(frutas)

```
